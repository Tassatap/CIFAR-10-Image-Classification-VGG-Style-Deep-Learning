#CIFAR-10 Image Classification — VGG-Style Deep Learning

CIFAR-10 is a widely used benchmark dataset consisting of 60,000 color images across 10 classes, such as airplanes, cars, birds, and animals. Despite its relatively small image size (32×32), the dataset presents a non-trivial challenge due to inter-class similarity and limited resolution.

To address this, the project adopts a VGG-style architecture, which is known for its simplicity and effectiveness in deep learning. The key idea behind VGG is the use of multiple stacked convolutional layers with small (3×3) filters, allowing the model to learn hierarchical feature representations — from low-level edges and textures to high-level object structures.

1️⃣ — Import Libraries

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

2️ split data and nomalize

```python
(x, y), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()
y, y_test = y.flatten(), y_test.flatten()
x = x / 255
x_test = x_test / 255
```

3️ plot data

```python
plt.imshow(x[11, :, :, :])
plt.show()
print(f"Class= {y[11]}")
```

4️⃣ Split the training set

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(x, y, test_size=0.2, random_state=11)
X_test = x_test

print("Shape of the datasets")
print("Training set   =", X_train.shape, y_train.shape)
print("Validation set =", X_valid.shape, y_valid.shape)
print("Test set       =", X_test.shape, y_test.shape)
```

5️⃣ Use one-hot encoding for y



```python
Y_train = tf.one_hot(y_train, 10)
Y_valid = tf.one_hot(y_valid, 10)
Y_test = tf.one_hot(y_test, 10)
```

6️⃣ creating model

```python

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, BatchNormalization, Dropout

model = Sequential([
    Input(shape=(32, 32, 3)),

    # Conv Block 1
    Conv2D(32, (3, 3), activation='relu', padding='same'),
    Conv2D(32, (3, 3), activation='relu', padding='same'),
    MaxPooling2D((2, 2)),

    # Conv Block 2
    Conv2D(64, (3, 3), activation='relu', padding='same'),
    Conv2D(64, (3, 3), activation='relu', padding='same'),
    MaxPooling2D((2, 2)),

    # Classifier
    Flatten(),
    Dense(256, activation='relu'),
    Dense(10, activation='softmax')
])

model.summary()

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

hist = model.fit(
    X_train, Y_train,
    batch_size=256,
    validation_data=(X_valid, Y_valid),
    epochs=50
)
```
